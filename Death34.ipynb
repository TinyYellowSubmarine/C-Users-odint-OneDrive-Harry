Построим круговую диаграмму используя библиотеку seaborn.

```python
#pie chart
death4 = death[(death["Country/Territory"] == 'Russia')]
pie, ax = plt.subplots (figsize = [8, 8])
data = death4[(death4["Year"] == 1990)]
labels = data.iloc[:,3:33].keys()
res = data.iloc[:,3:33].stack().reset_index(name="val").rename(columns={"level_1":"key"})
plt.pie (x = res.iloc[:,2], autopct = "%.1f%%", labels = labels, pctdistance = 0.75)
plt.title("Количество болезней в России на 1990 год", fontsize = 20)
```

```python
#pie chart
death4 = death[(death["Country/Territory"] == 'Russia')]
pie, ax = plt.subplots (figsize = [8, 8])
data = death4[(death4["Year"] == 2010)]
labels = data.iloc[:,3:33].keys()
res = data.iloc[:,3:33].stack().reset_index(name="val").rename(columns={"level_1":"key"})
plt.pie (x = res.iloc[:,2], autopct = "%.1f%%", labels = labels, pctdistance = 0.75)
plt.title("Количество болезней в России на 2010 год", fontsize = 20)
```

При необходимости можно вывести подобные диаграммы для любого другого массива данных. В данном случае были построены диаграммы по 2 годам. С  разницей в 20 лет.
